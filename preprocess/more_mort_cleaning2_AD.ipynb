```python
# Essentials
import pandas as pd
pd.set_option('display.max_columns', 100)
import csv
import numpy as np
import os

# Plots
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

# Stats
from scipy import stats
from scipy.stats import skew
```

```python
%%time 
# m2018 = pd.read_csv('data/mort_2018_filtered.csv')
# m2017 = pd.read_csv('data/mort_2017_filtered.csv')
m2018 = pd.read_csv('data/rawdata/mort_2018.csv', low_memory=False)
m2017 = pd.read_csv('data/rawdata/mort_2017.csv', low_memory=False)
m2016 = pd.read_csv('data/rawdata/mort_2016.csv', low_memory=False)
m2015 = pd.read_csv('data/rawdata/mort_2015.csv', low_memory=False)
m2014 = pd.read_csv('data/rawdata/mort_2014.csv', low_memory=False)
m2013 = pd.read_csv('data/rawdata/mort_2013.csv', low_memory=False)
m2012 = pd.read_csv('data/rawdata/mort_2012.csv', low_memory=False)
m2011 = pd.read_csv('data/rawdata/mort_2011.csv', low_memory=False)
```

```python
# replace blanks in manner_of_death with 0, change dtype to int64
m2018['manner_of_death'] = m2018['manner_of_death'].apply(pd.to_numeric, errors='coerce').fillna(0).astype('int64')
print('m2018 manner_of_death complete')
m2017['manner_of_death'] = m2017['manner_of_death'].apply(pd.to_numeric, errors='coerce').fillna(0).astype('int64')
print('m2017 manner_of_death complete')
m2016['manner_of_death'] = m2016['manner_of_death'].apply(pd.to_numeric, errors='coerce').fillna(0).astype('int64')
print('m2016 manner_of_death complete')
m2015['manner_of_death'] = m2015['manner_of_death'].apply(pd.to_numeric, errors='coerce').fillna(0).astype('int64')
print('m2015 manner_of_death complete')
m2014['manner_of_death'] = m2014['manner_of_death'].apply(pd.to_numeric, errors='coerce').fillna(0).astype('int64')
print('m2014 manner_of_death complete')
m2013['manner_of_death'] = m2013['manner_of_death'].apply(pd.to_numeric, errors='coerce').fillna(0).astype('int64')
print('m2013 manner_of_death complete')
m2012['manner_of_death'] = m2012['manner_of_death'].apply(pd.to_numeric, errors='coerce').fillna(0).astype('int64')
print('m2012 manner_of_death complete')
m2011['manner_of_death'] = m2011['manner_of_death'].apply(pd.to_numeric, errors='coerce').fillna(0).astype('int64')
print('m2011 manner_of_death complete')
```
